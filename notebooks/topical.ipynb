# Topical Chat

```python
import pandas as pd
import json
import tools
dataset = []
```

```python
filenames = ["test_freq", "test_rare", "valid_freq", "valid_rare", "train"]

for filename in filenames:
    with open(f"/workspaces/codespaces-jupyter/data/Topical-Chat/conversations/{filename}.json") as f:
        raw_json = f.read()
    json_data = json.loads(raw_json)
    for key in json_data:
        t = [i["message"] for i in json_data[key]["content"]]
        dataset.extend(t)
```

```python
data = []
for chank_num in range(0, len(dataset)//1000):
    date_sentces = tools.extract_date_roots(dataset[chank_num*1000:(chank_num+1)*1000])
    for i in date_sentces:
        if (i[1] == None):
            data.append([i[0], i[1], i[2], 0])
        else:
            data.append([i[0], i[1], i[2], 1])
    print(f"chunk {chank_num} completed")
data_df = pd.DataFrame(data, columns=["text", "root", "event", "label"])
data_df.to_csv("../processed_data/topicalchat_datesentences.csv")
```
